```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
try:
    with open('smartprix_real.html', 'r', encoding='utf-8') as file:
        html_content = file.read()
except FileNotFoundError:
    print("Error: The file 'smartprix_real.html' was not found.")
    exit()

soup = BeautifulSoup(html_content, 'html.parser')
```

```python
# The listings seem to be within a div with a specific class name.
# Inspecting the HTML, we'll look for a div that contains all the phone details.
# The class 'sm-product has-tag has-features has-actions' seems to be the main container for each phone.
phone_listings = soup.find_all('div', class_='sm-product has-tag has-features has-actions')

if not phone_listings:
    print("Error: Could not find any smartphone listings. The HTML structure may have changed.")
    exit()
```

```python
phone_listings
```

```python
model = []
price = []
rating = []
# ------------------- rest all below this separation are in the same container
sim = []
processor = []
ram = []
battery = []
display = []
camera = []
card = []
os = []

all_specs = []

for i in phone_listings:
    ul = i.find('ul', class_="sm-feat specs")
    if ul:
        li_tags = ul.find_all('li')
        specs = [li.get_text(strip=True) for li in li_tags]
        all_specs.append(specs)

for i in phone_listings:
    idx = 0
    try:
        model.append(i.find('h2').text.strip())
    except:
        model.append(np.nan)

    try:
        price.append(i.find('span', class_='price').text.strip())
    except: 
        price.append(np.nan)
    
    try: 
        rating.append(i.find('b').text.strip())
    except: 
        rating.append(np.nan)
    
    try: 
        sim.append(all_specs[idx][0])
    except: 
        sim.append(np.nan)
    
    try: 
        processor.append(all_specs[idx][1])
    except: 
        processor.append(np.nan)
    
    try: 
        ram.append(all_specs[idx][2])
    except: 
        ram.append(np.nan)
    
    try: 
        battery.append(all_specs[i][3])
    except: 
        battery.append(np.nan)
    
    try: 
        display.append(all_specs[idx][4])
    except: 
        display.append(np.nan)
    
    try: 
        camera.append(all_specs[idx][5])
    except: 
        camera.append(np.nan)
    
    try: 
        card.append(all_specs[idx][6])
    except: 
        card.append(np.nan)
    
    try: 
        os.append(all_specs[idx][7])
    except: 
        os.append(np.nan)
    idx+=1

df = pd.DataFrame({'model': model, 'price': price ,'rating': rating, 
    'sim': sim, 'processor': processor, 'ram': ram, 'battery': battery, 
    'display': display, 'camera': camera, 'card': card, 'os': os})
df
```
